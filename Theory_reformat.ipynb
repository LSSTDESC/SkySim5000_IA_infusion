```python
#Notebook used to plot measurements from the SkySim5000+IA mocks vs theory predictions
%pylab inline
import pylab as plt
import numpy as np
import treecorr
from astropy.io import fits

```

```python
PlotXipm = False
PlotRatio = False

#loop over cosmologies:
for ii in range(0,26):
    # Load the CCL predictions:
    #hdul = fits.open('theory/CCL/theo_prediction_KiDS-Legacy_cosmoSLICS_a1_p1.0_bta_1.0_a2_0.0.fits')
    hdul = fits.open('theory/CCL/KiDS-1000/theo_prediction_KiDS-1000_cosmo'+str(ii)+'_a1_p1.0_bta_1.0_a2_0.0.fits')
    #hdul = fits.open('theory/CCL/KiDS-Legacy/theo_prediction_KiDS-Legacy_cosmo'+str(ii)+'_a1_p1.0_bta_1.0_a2_0.0.fits')
    data=hdul[1].data
    NO_IA_plus_CCL=data['NO IA+'] # Xi + (no ia)
    NO_IA_minus_CCL=data['NO IA-'] # Xi - (no ia)
    GI_plus_CCL=data['GI+']        #GI + (TATT) 
    GI_minus_CCL=data['GI-']       #GI - (TATT)
    GI_NLA_plus_CCL=data['GI_NLA+']  #GI + (NLA),
    GI_NLA_minus_CCL=data['GI_NLA-']  #GI - (NLA)
    II_plus_CCL=data['II+']           #II + (TATT) 
    II_minus_CCL=data['II-']         #II -   (TATT)
    II_NLA_plus_CCL=data['II_NLA+']  #II +   (NLA)
    II_NLA_minus_CCL=data['II_NLA-']     #II -  (NLA)
    theta_CCL=data['theta']            # Angular separation(s) in arcmin
    z_combination=data['bin_comb']  #redshift bin combination 
    
    # Initialize:
    #KiDS-1000:
    Nbins=int(9);    Ntomo=int(5)
    #KiDS-Legacy, LSST,Euclid
    #Nbins=int(20);   Ntomo=int(6)
    
    Ntomocomb = int(Ntomo*(Ntomo+1)/2);
    counter=0
    xip=np.empty([Nbins,Ntomocomb])
    xim=np.empty([Nbins,Ntomocomb])
    xip_IA=np.empty([Nbins,Ntomocomb])
    xim_IA=np.empty([Nbins,Ntomocomb])
    xip_pureIA=np.empty([Nbins,Ntomocomb])
    xim_pureIA=np.empty([Nbins,Ntomocomb])
    xip_GI=np.empty([Nbins,Ntomocomb])
    xim_GI=np.empty([Nbins,Ntomocomb])
    IG_plus_CCL=np.empty([Ntomocomb,Nbins])
    IG_minus_CCL=np.empty([Ntomocomb,Nbins])
    IG_NLA_plus_CCL=np.empty([Ntomocomb,Nbins])
    IG_NLA_minus_CCL=np.empty([Ntomocomb,Nbins])
    
    # all:
    for bin1 in np.arange(Ntomo)+1:
        for bin2 in range (bin1,Ntomo+1):
    
            #print('Working on bins', bin1, bin2)
    
            #Ingesting theory (CCL)
            # Find the right column:
            for tomo in range(0,Ntomocomb):
                if(z_combination[tomo][0]==bin2 and z_combination[tomo][1]==bin1 ):
                    #print(z_combination[tomo][0], z_combination[tomo][1])
                    #print("Will use column",tomo)
                    break
    
            theta_CCL=data['theta']*60
            theta_CCL=theta_CCL[0,:]
            xip_CCL=NO_IA_plus_CCL[tomo,:]
            xim_CCL=NO_IA_minus_CCL[tomo,:]
    
            if(bin1==bin2):
                IG_plus_CCL[tomo,:] = GI_plus_CCL[tomo,:]
                IG_minus_CCL[tomo,:] = GI_minus_CCL[tomo,:]
    
                IG_NLA_plus_CCL[tomo,:] = GI_NLA_plus_CCL[tomo,:]
                IG_NLA_minus_CCL[tomo,:] = GI_NLA_minus_CCL[tomo,:]
    
            else:
                IG_plus_CCL[tomo,:] = np.zeros_like(GI_plus_CCL[tomo,:])
                IG_minus_CCL[tomo,:] = np.zeros_like(GI_minus_CCL[tomo,:])
    
            # Total signal:
            xip_IA_CCL=NO_IA_plus_CCL[tomo,:] +GI_plus_CCL[tomo,:] +IG_plus_CCL[tomo,:] + II_plus_CCL[tomo,:]
            xim_IA_CCL=NO_IA_minus_CCL[tomo,:]+GI_minus_CCL[tomo,:]+IG_minus_CCL[tomo,:] + II_minus_CCL[tomo,:]
    
            xip_IA_NLA_CCL=NO_IA_plus_CCL[tomo,:] +GI_NLA_plus_CCL[tomo,:] +IG_NLA_plus_CCL[tomo,:] + II_NLA_plus_CCL[tomo,:]
            xim_IA_NLA_CCL=NO_IA_minus_CCL[tomo,:]+GI_NLA_minus_CCL[tomo,:]+IG_NLA_minus_CCL[tomo,:] + II_NLA_minus_CCL[tomo,:]
    
    
            xip_II_CCL=II_plus_CCL[tomo,:]
            xim_II_CCL=II_minus_CCL[tomo,:]
            xip_II_NLA_CCL=II_NLA_plus_CCL[tomo,:]
            xim_II_NLA_CCL=II_NLA_minus_CCL[tomo,:]
    
            xip_GI_CCL=GI_plus_CCL[tomo,:]
            xim_GI_CCL=GI_minus_CCL[tomo,:]
            xip_GI_NLA_CCL=GI_NLA_plus_CCL[tomo,:]
            xim_GI_NLA_CCL=GI_NLA_minus_CCL[tomo,:]
    
            if(PlotXipm):
                #plot xi+:
        
                #CCL:
                plt.plot(theta_CCL, xip_CCL, 'k-', label='$\\xi+$, CCL')
                plt.plot(theta_CCL, xip_IA_CCL, 'r-')#, label='$\\xi+$, CCL-IA')
                plt.plot(theta_CCL, xip_II_CCL, 'g-')#, label='$\\xi+$, CCL-II')
                plt.plot(theta_CCL, -xip_GI_CCL, 'b-')#, label='$\\xi+$, CCL-II')
                plt.plot(theta_CCL, xip_IA_NLA_CCL, 'r--')#, label='$\\xi+$, CCL-IA')
                plt.plot(theta_CCL, xip_II_NLA_CCL, 'g--')#, label='$\\xi+$, CCL-II')
                plt.plot(theta_CCL, -xip_GI_NLA_CCL, 'b--')#, label='$\\xi+$, CCL-II')
        
        
                plt.loglog()
                plt.xlabel('$\\vartheta [arcmin]$', fontsize=16)
                plt.ylabel('$\\xi_+$', fontsize=16)
                plt.legend(loc='lower right', ncol=2, labelspacing=0.1)
                plt.xlim([0.3, 500])
                plt.ylim([1e-9, 1e-4])
                plt.show()
        
        
        
                #plot xi-        
        
                plt.plot(theta_CCL, xim_CCL, 'k-')#, label='$\\xi-$, CCL')
                plt.plot(theta_CCL, xim_IA_CCL, 'r-')#, label='$\\xi-$, CCL-IA')
                plt.plot(theta_CCL, xim_IA_NLA_CCL, 'r--')#, label='$\\xi-$, CCL-IA')
                plt.plot(theta_CCL, xim_II_CCL, 'g-')#, label='$\\xi-$, CCL-II')
                plt.plot(theta_CCL, -xim_GI_CCL, 'b-')#, label='$\\xi-$, CCL-II')       
                plt.plot(theta_CCL, xim_II_NLA_CCL, 'g--')#, label='$\\xi+$, CCL-II')
                plt.plot(theta_CCL, -xim_GI_NLA_CCL, 'b--')#, label='$\\xi+$, CCL-GI')
        
        
                plt.loglog()
                plt.xlabel('$\\vartheta [arcmin]$', fontsize=16)
                plt.ylabel('$\\xi_-$', fontsize=16)
                plt.legend(loc='lower right', ncol=2, labelspacing=0.1)
                plt.xlim([0.3, 500])
                plt.ylim([1e-9, 1e-4])
        
                plt.show()
        
            if(PlotRatio):
                        #plot xi+:
        
                #CCL:
                plt.plot(theta_CCL, xip_IA_CCL/xip_CCL -1, 'k-')#, label='$\\xi+$, CCL-IA')
                plt.plot(theta_CCL, xip_IA_NLA_CCL/xip_CCL -1, 'k--')#, label='$\\xi+$, CCL-IA')
                
        
        
                plt.semilogx()
                plt.xlabel('$\\vartheta [arcmin]$', fontsize=16)
                plt.ylabel('$\\xi_+/\\xi_+^{noIA} -1$', fontsize=16)
                plt.legend(loc='lower right', ncol=2, labelspacing=0.1)
                plt.xlim([0.3, 500])
                plt.ylim([-1.5, 0.2])
                plt.show()
        
        
        
                #plot xi-        
        
                plt.plot(theta_CCL, xim_IA_CCL/xim_CCL-1, 'k-')#, label='$\\xi-$, CCL-IA')
                plt.plot(theta_CCL, xim_IA_NLA_CCL/xim_CCL-1, 'k--')#, label='$\\xi-$, CCL-IA')
       
        
                plt.semilogx()
                plt.xlabel('$\\vartheta [arcmin]$', fontsize=16)
                plt.ylabel('$\\xi_-/\\xi_-^{noIA}-1$', fontsize=16)
                plt.legend(loc='lower right', ncol=2, labelspacing=0.1)
                plt.xlim([0.3, 500])
                plt.ylim([-1.5, 0.3])
        
                plt.show()
        
    
    
    
            counter +=1
    
    
    
    # Save in asciss-format files:
    
    # noIA:
    xip_flat = reshape(NO_IA_plus_CCL, [1,Nbins* Ntomocomb])
    xim_flat = reshape(NO_IA_minus_CCL, [1,Nbins* Ntomocomb])
    theta_flat = np.tile(theta_CCL, Ntomocomb)
    
    result = np.zeros([Nbins* Ntomocomb,2])
    for i in range(Nbins* Ntomocomb):
        result[i,0] = theta_flat[i]
        result[i,1] = xip_flat[0,i]
    
    #np.savetxt('theory/CCL/theo_xip_KiDS-Legacy_GG_cosmoSLICS_fid.dat',result)
    np.savetxt('theory/CCL/KiDS-1000/theo_xip_KiDS-1000_GG_cosmoSLICS_cosmo'+str(ii)+'.dat',result)
    #np.savetxt('theory/CCL/KiDS-Legacy/theo_xip_KiDS-Legacy_GG_cosmoSLICS_cosmo'+str(ii)+'.dat',result)
    
    result = np.zeros([Nbins* Ntomocomb,2])
    for i in range(Nbins* Ntomocomb):
        result[i,0] = theta_flat[i]
        result[i,1] = xim_flat[0,i]
    
    #np.savetxt('theory/CCL/theo_xim_KiDS-Legacy_GG_cosmoSLICS_fid.dat',result)
    np.savetxt('theory/CCL/KiDS-1000/theo_xim_KiDS-1000_GG_cosmoSLICS_cosmo'+str(ii)+'.dat',result)
    #np.savetxt('theory/CCL/KiDS-Legacy/theo_xim_KiDS-Legacy_GG_cosmoSLICS_cosmo'+str(ii)+'.dat',result)
    
    
    # IA:
    xip_flat = reshape(NO_IA_plus_CCL +GI_plus_CCL +IG_plus_CCL + II_plus_CCL, [1,Nbins* Ntomocomb])
    xim_flat = reshape(NO_IA_minus_CCL +GI_minus_CCL +IG_minus_CCL + II_minus_CCL, [1,Nbins* Ntomocomb])
    
    result = np.zeros([Nbins* Ntomocomb,2])
    for i in range(Nbins* Ntomocomb):
        result[i,0] = theta_flat[i]
        result[i,1] = xip_flat[0,i]
    
    #np.savetxt('theory/CCL/theo_xip_KiDS-Legacy_IA_cosmoSLICS_fid.dat',result)
    np.savetxt('theory/CCL/KiDS-1000/theo_xip_KiDS-1000_IA_cosmoSLICS_cosmo'+str(ii)+'.dat',result)
    #np.savetxt('theory/CCL/KiDS-Legacy/theo_xip_KiDS-Legacy_IA_cosmoSLICS_cosmo'+str(ii)+'.dat',result)

    result = np.zeros([Nbins* Ntomocomb,2])
    for i in range(Nbins* Ntomocomb):
        result[i,0] = theta_flat[i]
        result[i,1] = xim_flat[0,i]
    
    #np.savetxt('theory/CCL/theo_xim_KiDS-Legacy_IA_cosmoSLICS_fid.dat',result)
    np.savetxt('theory/CCL/KiDS-1000/theo_xim_KiDS-1000_IA_cosmoSLICS_cosmo'+str(ii)+'.dat',result)
    #np.savetxt('theory/CCL/KiDS-Legacy/theo_xim_KiDS-Legacy_IA_cosmoSLICS_cosmo'+str(ii)+'.dat',result)
    
    
    #NLA
    # IA:
    xip_flat = reshape(NO_IA_plus_CCL +GI_NLA_plus_CCL +IG_NLA_plus_CCL + II_NLA_plus_CCL, [1,Nbins* Ntomocomb])
    xim_flat = reshape(NO_IA_minus_CCL +GI_NLA_minus_CCL +IG_NLA_minus_CCL + II_NLA_minus_CCL, [1,Nbins* Ntomocomb])
    
    result = np.zeros([Nbins* Ntomocomb,2])
    for i in range(Nbins* Ntomocomb):
        result[i,0] = theta_flat[i]
        result[i,1] = xip_flat[0,i]
    
    #np.savetxt('theory/CCL/theo_xip_KiDS-Legacy_NLA_cosmoSLICS_fid.dat',result)
    np.savetxt('theory/CCL/KiDS-1000/theo_xip_KiDS-1000_NLA_cosmoSLICS_cosmo'+str(ii)+'.dat',result)
    #np.savetxt('theory/CCL/KiDS-Legacy/theo_xip_KiDS-Legacy_NLA_cosmoSLICS_cosmo'+str(ii)+'.dat',result)
    
    result = np.zeros([Nbins* Ntomocomb,2])
    for i in range(Nbins* Ntomocomb):
        result[i,0] = theta_flat[i]
        result[i,1] = xim_flat[0,i]
    
    #np.savetxt('theory/CCL/theo_xim_KiDS-Legacy_NLA_cosmoSLICS_fid.dat',result)
    np.savetxt('theory/CCL/KiDS-1000/theo_xim_KiDS-1000_NLA_cosmoSLICS_cosmo'+str(ii)+'.dat',result)
    #np.savetxt('theory/CCL/KiDS-Legacy/theo_xim_KiDS-Legacy_NLA_cosmoSLICS_cosmo'+str(ii)+'.dat',result)
```
