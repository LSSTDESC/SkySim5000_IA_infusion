```python
hist1, bin_edges1 = np.histogram(kappa1, bins=50, range=[-0.00001,0.00001])
hist2, bin_edges2 = np.histogram(kappa2, bins=50, range=[-0.00001,0.00001])
hist3, bin_edges3 = np.histogram(kappa3, bins=50, range=[-0.00001,0.00001])
hist4, bin_edges4 = np.histogram(kappa4, bins=50, range=[-0.00001,0.00001])
hist5, bin_edges5 = np.histogram(kappa5, bins=50, range=[-0.00005,0.00005])
hist6, bin_edges6 = np.histogram(kappa6, bins=50, range=[-0.00005,0.00005])
```

```python
plt.plot(bin_edges1[:-1], hist1,label='noIA')
plt.plot(bin_edges2[:-1], hist2,label='NLA')
plt.plot(bin_edges3[:-1], hist3,label='SC')
plt.plot(bin_edges4[:-1], hist4,label='deltaNLA')
plt.plot(bin_edges5[:-1], hist5,label='NLA, noisy')
plt.plot(bin_edges6[:-1], hist6,label='deltaNLA, noisy')
plt.axvline(x=0,color='grey',ls='--')
plt.legend(bbox_to_anchor=(1.0, 1), loc='upper left')
plt.xlabel(r'$\kappa$',fontsize=12)
plt.ylabel(r'$P(\kappa)$',fontsize=12)
```
